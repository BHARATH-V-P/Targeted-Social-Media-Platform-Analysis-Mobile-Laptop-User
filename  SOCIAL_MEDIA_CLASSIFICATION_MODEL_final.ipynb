### **<u>Skewness and Kurtosis Analysis<u>**

<u>Moderately Skewed Features<u>
- `'Yearly_avg_view_on_travel_page'`, `'total_likes_on_outstation_checkin_given'`, `'yearly_avg_Outstation_checkins'`, and `'total_likes_on_outofstation_checkin_received'` are moderately skewed towards the right.

<u>Highly Positive Skewed Features<u>
- `'Yearly_avg_comment_on_travel_page'`, `'montly_avg_comment_on_company_page'`, and `'Daily_Avg_mins_spend_on_traveling_page'` are highly positively skewed.

<u>Kurtosis Analysis<u>
- `'Yearly_avg_view_on_travel_page'`: Slightly platykurtic with flat peaks and light tails.
- `'total_likes_on_outstation_checkin_given'`: Leptokurtic with sharp peaks and heavy tails.
- `'yearly_avg_Outstation_checkins'`: Slightly platykurtic with light tails and flat peaks.
- `'Yearly_avg_comment_on_travel_page'`: Highly leptokurtic.
- `'total_likes_on_outofstation_checkin_received'`: The only feature near to normal distribution with kurtosis at 0.99.
- `'montly_avg_comment_on_company_page'`: Leptokurtic.
- `'Daily_Avg_mins_spend_on_traveling_page'`: Leptokurtic.

<u>Transformations and Handling Outliers<u>
- Highly skewed features like `'Yearly_avg_comment_on_travel_page'`, `'Daily_Avg_mins_spend_on_traveling_page'`, and `'montly_avg_comment_on_company_page'` may require transformations (e.g., quantile, log, sqrt, or scalar) to reduce skewness, particularly for linear models.
- Features with high kurtosis (e.g., `'Yearly_avg_comment_on_travel_page'`, `'montly_avg_comment_on_company_page'`, and `'Daily_Avg_mins_spend_on_traveling_page'`) indicate the presence of significant extreme outliers. These outliers should be handled based on the model's requirements.
- Features like `'Yearly_avg_view_on_travel_page'`, `'total_likes_on_outstation_checkin_given'`, and `'yearly_avg_Outstation_checkins'` require no transformations due to their closeness to normal distributions.

###  **<u>Outlier Detection Using Isolation Forest<u>**

___Detection Limitations___
- The Isolation Forest model has not removed all outliers from the numerical features.
- Outliers are detected based on the contamination rate set during the model's configuration.

___Anomaly Score Assignment___
- Initially, the model assigns an anomaly score to each data point based on its position (leaf) from the root of the tree.
- A shorter path from the root to a data point indicates fewer splits needed to isolate that point, suggesting it is an outlier.
- The data point with the shortest path receives the highest anomaly score.

___Isolation Process___
- Isolation Forest functions by:
  - Randomly picking a feature.
  - Creating a split at a value between the minimum and maximum values of the feature.
  - Continuing the splitting process until all data points are isolated, each represented by a leaf.

___Contamination Rate___
- A contamination rate of 0.05 considers the top 5% of isolated data points as outliers.
- A 5% contamination rate strikes a balance between outlier detection and avoiding overfitting, especially in skewed features.

___Column Characteristics___
- Isolation Forest operates independently on each feature.
- Columns with low variability or high data point density may have outliers that are not effectively detected.

### **<u>Ensembling using Voting Classsifer<u>**

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score,cohen_kappa_score,log_loss
import xgboost as xgb
import lightgbm as lgb

voting_type=[]
accuracy_scores=[]
cohen_kappa_scores=[]
logloss_scores=[]
model=[]
ml_model=[]


train_sets=[
           ('Laptop_model',X_train_laptop_1,y_train_laptop_1,X_test_laptop_1,y_test_laptop_1),
          
           ('Mobile_model',X_train_mobile_1,y_train_mobile_1,X_test_mobile_1,y_test_mobile_1)]

estimators=[('XGB',xgb.XGBClassifier(random_state=42,verbosity=0)),('LGB',lgb.LGBMClassifier(random_state=42,verbosity=-1)),
           ('RF',RandomForestClassifier(random_state=42))]
for j in train_sets:
    for i in ['hard','soft']:
        voting_type.append(i)
        model.append(j[0])
        vtg_clf=VotingClassifier(estimators=estimators,voting=i)
        vtg_clf.fit(j[1],j[2])
        accuracy_scores.append(accuracy_score(j[4],vtg_clf.predict(j[3])))
        cohen_kappa_scores.append(cohen_kappa_score(j[4],vtg_clf.predict(j[3])))
        if i=='soft':
            pred_prob_normalized=vtg_clf.predict_proba(j[3])/np.sum(vtg_clf.predict_proba(j[3]),axis=1,keepdims=True)
            logloss_scores.append(log_loss(j[4],pred_prob_normalized))
        else:
            logloss_scores.append('--')

voting_scores=pd.DataFrame({'Model':model,'Voting_type':voting_type,'Accuracy_score':accuracy_scores,
                           'Cohen_scores':cohen_kappa_scores,'Logloss':logloss_scores})
voting_scores
```
